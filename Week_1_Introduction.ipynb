## **3. [Drawing on Images](https://youtu.be/jN4DR_cxaEU)**

```python
# Draw a straight line
green = (0, 255, 0)
red = (0, 0, 255)
white = (255, 255, 255)
cv2.line(linus, (125, 300), (300, 300), green, thickness = 3)
cv2_imshow(linus)
```

```python
# Draw a circle 
cv2.circle(linus, (200, 150), 100, green, thickness = 2)
cv2_imshow(linus)
```

```python
# Draw a rectangle 
cv2.rectangle(linus, (100, 50), (300, 250), green)
cv2_imshow(linus)
```

```python
# Draw an ellipse
cv2.ellipse(linus, (210, 350), (100, 40), 0, 0, 360, green, -1)
cv2_imshow(linus)
```

```python
# Draw polylines
points = np.array([[[140, 325], [275, 325], [210, 375]]], np.int32)
cv2.polylines(linus, points, True, red, thickness = 2)
cv2_imshow(linus)
```

```python
# Put text
font = cv2.FONT_HERSHEY_COMPLEX_SMALL
cv2.putText(linus, 'Linus', (140, 360), font, 2, white, thickness = 3)
cv2_imshow(linus)
```

## **4. [Bitwise Operators](https://youtu.be/DCxmOIFWdRs)**

```python
img1 = cv2.imread('/content/drive/MyDrive/Computer Vision L1/images/img1.png')
img2 = cv2.imread('/content/drive/MyDrive/Computer Vision L1/images/img2.png')
```

```python
cv2_imshow(img1)
```

```python
cv2_imshow(img2)
```

```python
# bitwise and (only case: w+w result into white)
bit_and = cv2.bitwise_and(img1, img2)
cv2_imshow(bit_and)
```

```python
# bitwise or (cases: b+w or w+b result into white)
bit_or = cv2.bitwise_or(img1, img2)
cv2_imshow(bit_or)
```

```python
# bitwise xor
bit_xor = cv2.bitwise_xor(img1, img2)
cv2_imshow(bit_xor)
```

```python
# bitwise not (i)
bit_not1 = cv2.bitwise_not(img1)
cv2_imshow(bit_not1) 
```

```python
# bitwise not (ii)
bit_not2 = cv2.bitwise_not(img2)
cv2_imshow(bit_not2)
```

## **7. [Canny Edge Detection](https://www.youtube.com/watch?v=sRFM5IEqR2w)**
Reference: Medium [Article](https://towardsdatascience.com/canny-edge-detection-step-by-step-in-python-computer-vision-b49c3a2d8123) 

```python
Brando = cv2.imread('/content/drive/MyDrive/Computer Vision L1/images/Brando.jpg')
cv2_imshow(Brando)
```

```python
# Apply Gaussian Blur

Brando_Gaussian = cv2.GaussianBlur(Brando, (5, 5), 0)
cv2_imshow(Brando_Gaussian)
```

```python
Brando_Canny = cv2.Canny(Brando_Gaussian, 100, 200)
cv2_imshow(Brando_Canny)
```
